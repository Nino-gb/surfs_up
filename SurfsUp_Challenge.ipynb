```python
# Dependencies
import numpy as np
import datetime as dt
import pandas as pd
import matplotlib.pyplot as plt
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python
# Import the sqlalchemy extract function.
from sqlalchemy import extract

# 1. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 

june_results = []
june_results = session.query(Measurement.date, Measurement.tobs).filter(extract("month", Measurement.date)==6).all()
print(june_results)
```

```python
# 2. Convert the June temperatures to a list.
june_temp = list(june_results)
june_temp
```

```python
# 3. Create a DataFrame from the list of temperatures for the month of June. 
june_temp_df = pd.DataFrame(june_results, columns=['date',' June Temps'])
june_temp_df
```

```python
# 4. Calculate and print out the summary statistics for the June temperature DataFrame.
june_temp_df.describe()
```

## D2: Determine the Summary Statistics for December

```python
# 6. Write a query that filters the Measurement table to retrieve the temperatures for the month of December.
dec_results = []
dec_results = session.query(Measurement.date, Measurement.tobs).filter(extract("month", Measurement.date)==12).all()
print(dec_results)
```

```python
# 7. Convert the December temperatures to a list.
dec_temp = list(dec_results)
dec_temp
```

```python
# 8. Create a DataFrame from the list of temperatures for the month of December. 
dec_temp_df = pd.DataFrame(dec_results, columns=['date',' December Temps'])
dec_temp_df
```

```python
# 9. Calculate and print out the summary statistics for the Decemeber temperature DataFrame.
dec_temp_df.describe()
```

```python
# Compare counts for D1 and D2 results
compare_results= pd.concat([dec_temp_df, june_temp_df])
compare_results.count()
```

```python
# Compare your summary statistics of D1 and D2
# Create DF for june a december statistics
june_statistics = june_temp_df.describe()
dec_statistics = dec_temp_df.describe()
```

```python
# Joint june_statistics with dec_statistics
statistics = pd.concat([june_statistics, dec_statistics],axis = 1)
statistics
```

```python
# Drop Count row
statistics = statistics.drop(statistics.index[0])
statistics
```

```python
# Create Bar graphic with statistics
statistics.plot.barh(figsize=(12,5))
plt.xlabel("Temperature")
plt.ylabel("Statistics($)")
plt.title(" June and December Comparative Statics")
```

```python
# Create a query that filters the Measurement table to retrieve the precipitations for the month of December
dec_prcp = []
dec_prcp = session.query(Measurement.date, Measurement.prcp).filter(extract("month", Measurement.date)==12).all()
print(dec_prcp)
```

```python
# Convert the December precipitation to a list.
dec_prcp_list = list(dec_prcp)
dec_prcp_list
```

```python
# Convert December precipitation list in DataFrame
dec_prcp_list = pd.DataFrame(dec_prcp_list, columns=['Date',' December Precipitation'])
dec_prcp_list
```

```python
# Create a variable that call December precipitation average by year
dec_prcp_years=dec_prcp_list.groupby(pd.PeriodIndex(dec_prcp_list['Date'], freq="Y"))[' December Precipitation'].mean()
```

```python
# Convert December annual precipitation variable in DataFrame
dec_prcp_years =pd.DataFrame(dec_prcp_years)
dec_prcp_years
```

```python
# Create a query that filters the Measurement table to retrieve the precipitations for the month of June
jun_prcp = []
jun_prcp = session.query(Measurement.date, Measurement.prcp).filter(extract("month", Measurement.date)== 6).all()
print(jun_prcp)
```

```python
# Convert the June precipitation to a list.
jun_prcp_list = list(jun_prcp)
jun_prcp_list
```

```python
# Convert June precipitation variable in DataFrame
jun_prcp_list = pd.DataFrame(jun_prcp_list, columns=['Date',' June Precipitation'])
jun_prcp_list
```

```python
# Create a variable that call june precipitation average by year
june_prcp_years= jun_prcp_list.groupby(pd.PeriodIndex(jun_prcp_list['Date'], freq="Y"))[' June Precipitation'].mean()
```

```python
# # Convert December annual precipitation variable in DataFrame
june_prcp_years =pd.DataFrame(june_prcp_years)
june_prcp_years
```

```python
# Joint december and june annual precipitation average Dataframe
years_prpc = pd.concat([june_prcp_years, dec_prcp_years],axis = 1)
years_prpc
```

```python
# Creat line graph statistics for June and December Annual Precipitation average
years_prpc.plot(figsize=(12,5))
plt.xlabel("Year")
plt.ylabel('Precipitation')
plt.title(" June and December Annual Precipitation Average")
```
